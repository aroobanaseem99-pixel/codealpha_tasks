```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re

nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
print("✅ Libraries loaded!")
```

```python
faqs = [
    {"q": "What is Artificial Intelligence?",
     "a": "Artificial Intelligence (AI) is the simulation of human intelligence by machines. It includes learning, reasoning, and self-correction."},
    {"q": "What is machine learning?",
     "a": "Machine Learning is a subset of AI where systems learn from data and improve automatically without being explicitly programmed."},
    {"q": "What is deep learning?",
     "a": "Deep Learning uses neural networks with many layers to learn complex patterns from large amounts of data."},
    {"q": "What is Python?",
     "a": "Python is a popular high-level programming language widely used in AI, data science, and web development."},
    {"q": "What is a neural network?",
     "a": "A neural network is a system inspired by the human brain, made of layers of connected nodes that process information."},
    {"q": "What is natural language processing?",
     "a": "NLP is a field of AI that helps computers understand, interpret, and generate human language."},
    {"q": "What is computer vision?",
     "a": "Computer Vision is an AI field that enables machines to interpret and understand visual information from images and videos."},
    {"q": "What is ChatGPT?",
     "a": "ChatGPT is an AI chatbot developed by OpenAI that uses large language models to have conversations with users."},
    {"q": "How does AI work?",
     "a": "AI works by processing large amounts of data, finding patterns, and using those patterns to make decisions or predictions."},
    {"q": "What is data science?",
     "a": "Data Science is a field that uses statistics, programming, and AI techniques to extract insights from data."},
    {"q": "What is a chatbot?",
     "a": "A chatbot is an AI-powered program that can simulate conversation with human users through text or voice."},
    {"q": "What is reinforcement learning?",
     "a": "Reinforcement learning is a type of ML where an agent learns by receiving rewards or penalties based on its actions."},
    {"q": "What is supervised learning?",
     "a": "Supervised learning is when a model is trained on labeled data — each input has a known correct output."},
    {"q": "What is unsupervised learning?",
     "a": "Unsupervised learning finds hidden patterns in data without labeled examples or predefined outputs."},
    {"q": "What is overfitting?",
     "a": "Overfitting happens when a model learns training data too well and performs poorly on new, unseen data."},
]

print(f"✅ {len(faqs)} FAQs loaded!")
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    stop_words = set(stopwords.words('english'))
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    return ' '.join(tokens)

questions = [preprocess(f['q']) for f in faqs]
answers = [f['a'] for f in faqs]
original_questions = [f['q'] for f in faqs]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(questions)

def get_answer(user_input, threshold=0.1):
    processed = preprocess(user_input)
    user_vec = vectorizer.transform([processed])
    similarities = cosine_similarity(user_vec, tfidf_matrix)[0]
    best_idx = np.argmax(similarities)
    best_score = similarities[best_idx]
    if best_score < threshold:
        return None, None, 0
    return original_questions[best_idx], answers[best_idx], round(best_score * 100, 1)

print("✅ NLP Engine ready!")
```

```python
print("=" * 55)
print("       🤖 CodeAlpha AI FAQ Chatbot")
print("=" * 55)
print("Type your question and press Enter.")
print("Type 'quit' to exit.\n")

chat_history = []

while True:
    user_input = input("You: ").strip()
    if not user_input:
        continue
    if user_input.lower() in ['quit', 'exit', 'bye']:
        print("Bot: Goodbye! Have a great day! 👋")
        break

    matched_q, answer, confidence = get_answer(user_input)

    if answer:
        print(f"\nBot: {answer}")
        print(f"     [Matched: '{matched_q}' | Confidence: {confidence}%]\n")
        chat_history.append({"user": user_input, "bot": answer})
    else:
        print("\nBot: I'm sorry, I don't have an answer for that. Please try rephrasing your question.\n")
```

```python
# Automatic test — manually check kar sakte ho
test_questions = [
    "tell me about AI",
    "what is ml",
    "how do neural nets work",
    "explain deep learning",
    "what is python used for"
]

print("🧪 Auto Testing Chatbot:\n")
print("-" * 50)
for q in test_questions:
    matched_q, answer, conf = get_answer(q)
    print(f"Q: {q}")
    if answer:
        print(f"A: {answer[:80]}...")
        print(f"   Confidence: {conf}%")
    else:
        print("A: No match found.")
    print("-" * 50)
```
